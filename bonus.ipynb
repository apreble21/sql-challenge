```python
#import needed modules
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import psycopg2
from sqlalchemy import create_engine
from config import password as pw
```

```python
#connect to postgres database
engine = create_engine(f'postgresql://postgres:{pw}@localhost:5432/Pewlett_Hackard')
connection = engine.connect()
```

```python
#import salaries data from postgres
salaries = pd.read_sql("SELECT * FROM salaries", connection)
#inspect data to determine if null values exist
salaries.describe()
```

```python
#create plot variable for easier manuipulation
ax = salaries.hist(column='salary', bins=25, grid=False, figsize=(12,8), color='#00cc66', zorder=2, rwidth=0.9)

ax = ax[0]
for x in ax:

    # Despine & switch off ticks for cleaner visualization
    x.spines['right'].set_visible(False)
    x.spines['top'].set_visible(False)
    x.spines['left'].set_visible(False)
    x.tick_params(axis="both", which="both", bottom="on", top="off", labelbottom="on", left="off", right="off", labelleft="on")

    # Add dashed horizontal axis lines for readability
    vals = x.get_yticks()
    for tick in vals:
        x.axhline(y=tick, linestyle='dashed', alpha=0.4, color='grey', zorder=1)

    # Set title, x-axis label, & y-axis label
    x.set_title("Salary Distribution at Pewlett Hackard",weight='bold', size=16)
    x.set_xlabel("Salary", labelpad=20, size=14)
    x.set_ylabel("Employee Count at Salary Range ", labelpad=20, size=14)
#save histogram as png
plt.savefig("SalaryDistribution.png", bbox_inches='tight') 
```

```python
#import titles data from postgres
titles = pd.read_sql("SELECT * FROM titles", connection)
titles.head()
```

```python
#import employees data from postgres
employees = pd.read_sql("SELECT emp_no, emp_title_id FROM employees", connection)
#rename column emp_title_id to match coordinating column in titles
employees = employees.rename(columns={'emp_no': 'emp_no', 'emp_title_id': 'title_id' })
employees.head()
```

```python
#merge titles & employees 
emp_title_merge = pd.merge(titles, employees, on='title_id', how='inner')
emp_title_merge.head()
```

```python
#merge last database with salaries
emp_title_salary = pd.merge(emp_title_merge, salaries, on='emp_no', how='inner')
emp_title_salary.head()
```

```python
#group by title, find average of each title, sort, and reset index
group_by_title = emp_title_salary.groupby('title')
salary_mean = pd.DataFrame(group_by_title['salary'].mean())
salary_mean_ordered = salary_mean.sort_values(['salary'])
salary_mean_clean = salary_mean_ordered.reset_index()
salary_mean_clean
```

```python
#define values for graph
x_axis = salary_mean_clean['title']
ticks = list(x_axis)
y_axis = salary_mean_clean['salary']
#create plot variable for easier manuipulation
ax_2 = salary_mean_clean.plot(kind='barh', figsize=(7, 3), color='#00cc66')

# Despine for cleaner visualization 
ax_2.spines['right'].set_visible(False)
ax_2.spines['top'].set_visible(False)
ax_2.spines['left'].set_visible(False)
ax_2.spines['bottom'].set_visible(False)

# Set title, x-axis label, & y-axis label, y-tick labels, & label values
ax_2.set_title("Average Salary at Pewlett Hackard", pad=20, weight='bold', size=16)
ax_2.set_xlabel("Average Salary", labelpad=20, weight='bold', size=12)
ax_2.set_ylabel("Job Title", labelpad=20, weight='bold', size=12)
ax_2.set_yticklabels(ticks)
ax_2.set_xlim(45000, 60000)
for x in ax_2.patches:
    ax_2.annotate("%.2f" % x.get_width(), (x.get_x() + x.get_width(), x.get_y()), 
                  xytext=(1, 1), textcoords='offset points')
# Save bar chart as png     
plt.savefig("AverageSalary.png", bbox_inches='tight')    
```
